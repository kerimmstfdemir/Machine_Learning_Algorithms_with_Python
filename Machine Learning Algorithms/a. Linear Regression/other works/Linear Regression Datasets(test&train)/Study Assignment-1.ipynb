###### Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

###### Import Data

```python
dirty_data = pd.read_csv("datasets/train.csv")
data = dirty_data.dropna()
```

###### Plotting Data

```python
plt.scatter(data.x,data.y)
plt.xlabel("x value")
plt.ylabel("y value")
plt.rcParams["figure.figsize"] = [25,15]
plt.show()
```

###### Linear Regression

```python
linear_reg = LinearRegression()
x = data.x.values.reshape(-1,1)
y = data.y.values.reshape(-1,1)
linear_reg.fit(x,y)
```

###### Prediction

```python
b0 = linear_reg.predict([[0]])
print("b0 =",b0)
b0_ = linear_reg.intercept_
print("b0_ =",b0_)
```

```python
b1 = linear_reg.coef_
print("b1 =",b1)
```

```python
linear_reg.predict([[25]])
```

###### Line Visulation

```python
array = np.array(range(0,101)).reshape(-1,1)
y_head = linear_reg.predict(array)
plt.scatter(x,y)
plt.plot(array,y_head, color = "red")
plt.show()
```
