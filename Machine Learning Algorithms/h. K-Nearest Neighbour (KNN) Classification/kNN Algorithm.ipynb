```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("kNN dataset.csv")
```

```python
dataset.head()
```

```python
dataset.drop(["id","Unnamed: 32"],axis = 1, inplace = True)
dataset.tail()
# Malignant = 'M'
# Benign    = 'B'
```

```python
M = dataset[dataset.diagnosis == "M"]
B = dataset[dataset.diagnosis == "B"]
```

```python
M.info()
B.info()
```

```python
# Scatter Plot
plt.scatter(M.radius_mean,M.texture_mean,color = 'red', label = "Malignant", alpha = 0.3)
plt.scatter(B.radius_mean,B.texture_mean,color = 'green', label = "Benign", alpha = 0.3)
plt.xlabel("radius_mean")
plt.ylabel("texture_mean")
plt.legend()
plt.show()
```

```python
dataset.diagnosis = [1 if each == "M" else 0 for each in dataset.diagnosis]
y = dataset.diagnosis.values
x_data = dataset.drop(["diagnosis"],axis = 1)
```

```python
#Normalization
x = ((x_data - np.min(x_data)) / ((np.max(x_data)) - np.min(x_data)))
```

```python
#Train-Test Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3, random_state =1)
```

```python
#kNN Model
from sklearn.neighbors import KNeighborsClassifier
kNN = KNeighborsClassifier(n_neighbors=3)  #n_neighbors = k value
kNN.fit(x_train,y_train)
prediction = kNN.predict(x_test)
```

```python
print("kNN Score (k = {}) : {}".format(3,kNN.score(x_test,y_test)))
```

```python
#Finding best k value
score_list = []
for each in range(1,15):
    kNN2 = KNeighborsClassifier(n_neighbors=each)
    kNN2.fit(x_train,y_train)
    score_list.append(kNN2.score(x_test,y_test))
    
plt.plot(range(1,15),score_list)
plt.xlabel("k Values")
plt.ylabel("Accuracy")
plt.show()
```

# Conclusion
* If we select our "k value" as 6,7,8,9,10 or 12, we can achive the best accuracy.
